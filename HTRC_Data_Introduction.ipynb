## Most Common Tokens in Your Volume (By Part of Speech)

**Function:** View the top 15 most common tokens in your volume as a table and bar chart. Optionally, you can display only the top words of a given part of speech.

**Inputs:**

*   ```POS``` (Optional): Specify the part of speech of the words you would like to display. Tokens in HTRC are tagged using the [Penn Treebank scheme](https://www.ling.upenn.edu/courses/Fall_2003/ling001/penn_treebank_pos.html).  For example, if you would like to view the most frequent singular proper nouns, enter ```"NNP"```.
*   ```POS_TYPE``` (Optional): Write the part of speech you are examining as you would like it to appear in the title of the bar chart (for example, "Proper Noun").


````python
# Enter your inputs here (make sure your part of speech tag is inside quotation marks):
POS = "NNP"
POS_TYPE = "Proper Noun"
````

````python
# Run this cell after setting your inputs:
tokenlist = WORKING_VOL.tokenlist(case=False, section="body", pages=False)
if POS == "None":
  toptokens = tokenlist.loc[slice(None), slice(None), slice(None), ].sort_values('count', ascending=False).reset_index(level=[0, 2], drop=True).head(15)
  toptokens = pd.DataFrame({"token": toptokens.index, "count" : toptokens["count"]}).reset_index(drop=True)
  toptokens.plot(kind="bar", x="token", xlabel= POS_TYPE, ylabel="Count", legend=False,
                          title=f'Most Common {POS_TYPE + "s"} in {WORKING_VOL_TITLE}', rot=45, figsize=(12,8))
  print(toptokens)

else:
  try:
    toptokens = tokenlist.loc[slice(None), slice(None), POS, ].sort_values('count', ascending=False).reset_index(level=[0], drop=True).head(15)
    toptokens = pd.DataFrame({"token": toptokens.index, "count" : toptokens["count"]}).reset_index(drop=True)
    toptokens.plot(kind="bar", x="token", xlabel= POS_TYPE, ylabel="Count", legend=False,
                              title=f'Most Common {POS_TYPE + "s"} in {WORKING_VOL_TITLE}', rot=45, figsize=(12, 8))
    print(toptokens)
  except KeyError:
    print("Oops! Please check the Penn Treebank table to make sure your are entering a valid part of speech label.")
````

## Charting Occurrences of a Word (or Words) in Your Volume

**Function**: View how often a word occurs on each page of your book. Enter multiple words to compare them.

**Inputs**:


*   ```WORDS``` (Required): The word or words whose occurrences you would like to visualize. All words (including names) should be written in lowercase. Enter them inside the brackets in quotation marks and separate them with commas
*   ```START_PAGE``` (Optional): The page number where you would like your graph to start.
*   ```END_PAGE``` (Optional): The page number where you would like your graph to end.



````python
# Enter your inputs here:
WORDS = ["sauron"]
START_PAGE = None
END_PAGE = None
````

````python
# Run this cell after setting your inputs:
linecount = 0
tokenlist = WORKING_VOL.tokenlist(case=False, section="body", pos=False)
try:
  ax = tokenlist.loc[(slice(None), slice(None), WORDS[0]), ].reset_index(level=[1, 2], drop=True).plot(xlim=(START_PAGE, END_PAGE), xlabel="Page Number", ylabel="Number of Tokens",
                                                                                                       title=f"Occurrences of Tokens in {WORKING_VOL_TITLE}", figsize=(12, 8), ls=linestyles[0])
except:
  print(f'Oops! Looks like {WORDS[0]} is not occur in this book.')
if len(WORDS) > 1:
  for word in WORDS[1:]:
    linecount += 1
    try:
      tokenlist.loc[(slice(None), slice(None), word), ].reset_index(level=[1, 2], drop=True).plot(ls=linestyles[linecount % 4], xlabel="Page Number", ax=ax)
    except:
      print(f'Oops! Looks like {word} does not occur in this book.')
ax.legend(WORDS)
````
